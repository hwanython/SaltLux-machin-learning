```python
##########################
# 라이브러리 사용
import tensorflow as tf
import pandas as pd
```

```python
###########################
# 1.과거의 데이터를 준비합니다.
파일경로 = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/boston.csv'
보스턴 = pd.read_csv(파일경로)
 
# 종속변수, 독립변수
독립 = 보스턴[['crim', 'zn', 'indus', 'chas', 'nox', 
            'rm', 'age', 'dis', 'rad', 'tax', 'ptratio', 'b', 'lstat']]
종속 = 보스턴[['medv']]
print(독립.shape, 종속.shape)
```

```python
# 모델을 준비합니다. 

X = tf.keras.Input(shape=[13])
H = tf.keras.layers.Dense(5, activation='swish')(X)
Y = tf.keras.layers.Dense(1)(H)
model = tf.keras.Model(X, Y)
model.compile(loss='mse')
model.summary()
```

```python
# 모델을 준비합니다. 

X = tf.keras.Input(shape=[13])
H = tf.keras.layers.Dense(8, activation='swish')(X)
H = tf.keras.layers.Dense(8, activation='swish')(H)
H = tf.keras.layers.Dense(8, activation='swish')(H)
Y = tf.keras.layers.Dense(1)(H)
model = tf.keras.Model(X, Y)
model.compile(loss='mse', optimizer='adam')
model.summary()
```

```python
model.fit(독립, 종속,epochs =100)
```

# Batch Normalization

```python
# 모델을 준비합니다. 

X = tf.keras.Input(shape=[13])
# H = tf.keras.layers.Dense(8, activation='swish')(X)
# H = tf.keras.layers.Dense(8, activation='swish')(H)

#H = tf.keras.layers.Dense(8, activation='swish')(H)
H = tf.keras.layers.Dense(8)(X)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation("relu")(H)
H = tf.keras.layers.Dense(8)(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation("relu")(H)
H = tf.keras.layers.Dense(8)(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation("relu")(H)

Y = tf.keras.layers.Dense(1)(H)
model = tf.keras.Model(X, Y)
model.compile(loss='mse')
model.summary()
```

```python
model.fit(독립, 종속, epochs=500)
```

```python
# 데이터 준비하기
x_input = pd.DataFrame([
    [1, 2], [1, 5], [2, 2], [2, 9], [3, 1], [3, 7], [4, 3], [4, 8], [5, 8], [5, 3], 
    [6, 5], [6, 9], [7, 3], [7, 1], [8, 5], [8, 8], [9, 4], [9, 3],
])
x_input
```

```python
y_label = pd.DataFrame(x_input.sum(axis=1))
y_label
```

```python
X = tf.keras.Input(shape=[2])

H = tf.keras.layers.Dense(8)(X)
# H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation("relu")(H)

H = tf.keras.layers.Dense(8)(H)
# H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation("relu")(H)

H = tf.keras.layers.Dense(8)(H)
# H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation("relu")(H)

H = tf.keras.layers.Dense(8)(H)
# H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation("relu")(H)

H = tf.keras.layers.Dense(8)(H)
# H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation("relu")(H)

H = tf.keras.layers.Dense(8)(H)
# H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation("relu")(H)


Y = tf.keras.layers.Dense(1)(H)
model = tf.keras.Model(X, Y)
model.compile(loss='mse', optimizer="adam")
model.summary()
```

```python
model.fit(x_input, y_label, epochs=1000, verbose = 0)
model.fit(x_input, y_label, epochs=10, verbose = 1)
```
