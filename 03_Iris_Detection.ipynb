## 세번째 딥러닝 - 아이리스 품종 분류

아이리스: https://raw.githubusercontent.com/blackdew/ml-tensorflow/master/data/csv/iris.csv

```python
###########################
# 라이브러리 사용
import tensorflow as tf
import pandas as pd
 
```

```python
###########################
# 1.과거의 데이터를 준비합니다.
파일경로 = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/iris.csv'
아이리스 = pd.read_csv(파일경로)
아이리스.head()
```

```python
# 원핫인코딩
onehot = pd.get_dummies(아이리스)
onehot.head()
```

```python
독립 = onehot[['꽃잎길이', '꽃잎폭', '꽃받침길이', '꽃받침폭']]
종속 = onehot[['품종_setosa', '품종_versicolor', '품종_virginica']]
print(독립.shape, 종속.shape)
```

```python
# 모델을 준비합니다. 
X = tf.keras.Input(shape=[4])
Y = tf.keras.layers.Dense(3, activation='softmax')(X)
model = tf.keras.Model(X, Y)
model.compile(loss="categorical_crossentropy", metrics='accuracy')
model.summary()
```

```python
# 모델에 데이터를 학습합니다. 
model.fit(독립, 종속, epochs=100)
```

```python
테스트 = 독립.sample(10)
테스트
```

```python
# 모델을 이용합니다. 
model.predict(테스트)
```
